## 01-Traits
* A trait in Scala is a reusable unit of behavior that can contain methods and fields.
* It is similar to an interface in Java but more powerful because it can include method implementations.  

Traits are mainly used to achieve:
* ✅ Code reuse
* ✅ Multiple inheritance
* ✅ Behavior composition

**🔑 Key Characteristics**  
* Declared using the trait keyword
* Cannot have constructor parameters (Scala 2 style)
* Classes use extends or with to mix them in
* Can contain abstract and concrete methods
* Support multiple inheritance

```scala
trait logger{
    def log(msg: String): Unit = println(msg) // concreate method: have defination and implementation (body)
}

class Services extends logger

val s1 = new Services
s1.log("Hello this is first log")
```

```scala
trait car{
    def color(Cname: String): String // abstract method: have defination only
}

trait airBagInd{
    def airBagFun(indicatorVal: String): String // Abstract method: declaration only (no implementation)”
}

// multiple inheritance with trait
class Alto extends car with airBagInd{
    def color(Cname: String): String = Cname+" color"
    def airBagFun(indicatorVal: String) = indicatorVal + " air bag"
}

val a1 = new Alto
println(s"${a1.color("White")} Alto car with ${a1.airBagFun("No")}")
```

When we say a sealed trait can only be extended in the same file, it means:   

👉 All classes/objects that extend that trait must be written in the exact same .scala source file.  

* Not the same package
* Not the same folder
* Not the same project
* ✔ Same physical file

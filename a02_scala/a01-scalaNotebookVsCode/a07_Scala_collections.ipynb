## 01-Scala-List
* **Immutable:** Operations on a list return a new list rather than modifying the original.
* **Recursive Structure:** A list is either an empty list (Nil) or a "Cell" containing an element and a reference to the rest of the list.
* **Homogeneous:** All elements in a list must be of the same type (e.g., List[Int]).

```scala
val variable_name: List[type] = List(item1, item2, item3)
or
val variable_name = List(item1, item2, item3)
```

````scala
// defining the list
val fruitsList = List("apple", "banana", "papaya", "orange")

// other way to define list
val num = 1::2::3::Nil

// empty list
val emptyList = List()
````

### Operation-on-List
* `head:` Returns the first element of the list.
* `tail:` Returns everything except the first element.
* `isEmpty:` Returns true if the list has no elements.

````scala
// head => returns the first item of list
fruitsList.head

// tail => returns the list containing all element except the head (first item)
fruitsList.tail

// isEmpty => returns true if the list is empty
emptyList.isEmpty
````

````scala
val list1: List[String]=List("Plain Donut","Strawberry Donut","Chocolate Donut")
````

````scala
// access the element of list
list1(0)
list1(1)
````

````scala
// append the item in list
val list2: List[String] = list1 :+ "Vanilla Donut"
````

````scala
// prepend the elements at the front of immutable list using +:
val list3: List[String] = "Vanilla Donut" +: list1
````

````scala
// add two immutable list using :::
val list4: List[String] = list1 ::: list2
````

````scala
// add two immutable list using ::
val list5: List[Any] = list1 :: list2
````

### Common-Transfermation
| Method   | Description                                   | Example                                   |
|----------|-----------------------------------------------|-------------------------------------------|
| map      | Applies a function to every element.          | `List(1, 2).map(_ * 2) -> List(2, 4)`     |
| filter   | Keeps elements that match a predicate.        | `List(1, 4).filter(_ > 2) -> List(4)`     |
| flatMap  | Maps and then flattens the result.            | Useful for nested structures.             |
| foldLeft | Reduces the list to a single value from left. | `List(1, 2, 3).foldLeft(0)(_ + _) -> 6`   |


````scala
val numList: List[Int] = List(1,2,3,4,5,6,7,8,9)
````

````scala
// map to calculate the square
val sqNumList: List[Int] = numList.map((x:Int) => x*x)
````

````scala
// multiplying with 2 using wildcard _
val sqNumList01 = numList.map(_*2)
````

````scala
// filter the num greater than 5
val numGreaterThan5 = numList.filter((f: Int) => f > 5)
````

````scala
// same thing can be written using wildcard
val numGreaterThan4 = numList.filter(_>4)
````

## 02-Scala-Map
* Map is a collection of key-value pairs. it is similar to dictionary. 
* Keys are always unique while values need not be unique. 
* Key-value pairs can have any data type. However, data type once used for any key and value must be consistent throughout. 

**Maps are classified into two types:**   
mutable and immutable. By default Scala uses `immutable` Map. In order to use `mutable` Map, we must import `scala.collection.mutable.Map` class explicitly.

### Syntax
```scala
// Immutable
variable = Map(key_1 -> value_1, key_2 -> value_2,
 key_3 -> value_3, ....)

// Mutable
variable = scala.collection.mutable.Map(key_1 -> value_1, 
key_2 -> value_2, key_3 -> value_3, ....)
```

### Operations on a Scala Map
There are three basic operations we can carry out on a Map:
* **keys:** In Scala Map, This method returns an iterable containing each key in the map.
* **values:** Value method returns an iterable containing each value in the Scala map.
* **isEmpty:** This Scala map method returns true if the map is empty otherwise this returns false.

````scala
// creating Map for student(name, age)
val student: Map[String, Int]= Map(
    "Ram" -> 29,
    "Vishva" -> 28,
    "Shyam" -> 25
)
````

````scala
// Accessing Values Using Keys
val ram = student("Ram")
````

````scala
// Accessing the key which is not present in the Map
val amit = student("Amit") // since Amit is not present we are getting the key not found error we can handle this lets see below
````

````scala
// Accessing the key which is not present in the Map using contains
val amit = if(student.contains("Amit")) student("Amit") else 0
````

````scala
// Accessing the key which is not present in the Map using contains
val amit = if(student.contains("Vishva")) student("Vishva") else 0
````

### Updating the values
* If we try to update value of an immutable Map, Scala outputs an error. 
* On the other hand, any changes made in value of any key in case of mutable Maps is accepted. Example: Updating immutable Map:
`error: value update is not a member of scala.collection.immutable.Map[String, Int]`

````scala
// create mutable Map and update value
val studentMutable: scala.collection.mutable.Map[String, Int]= scala.collection.mutable.Map(
    "Ram" -> 29,
    "Vishva" -> 28,
    "Shyam" -> 25
)
````

````scala
studentMutable("Vishva")=30
````

````scala
studentMutable("Vishva")
````

````scala
// adding the key, value in Map
studentMutable += ("Radha" -> 25, "Anjali" -> 22)
````

````scala
studentMutable
````

````scala
// deleting key-value in Map | Ram key and its value will remove from the Map
studentMutable -= "Ram"
````

### Iteration in a Map
* Key-value pair corresponds to a tuple with two elements. 
* Therefore, while performing iteration loop variable needs to be a pair.

````scala
val studentMut = scala.collection.mutable.Map(
    "Ram" -> 29,
    "Vishva" -> 28,
    "Shyam" -> 25
)

for((i, j) <- studentMut){
    print(s"key: $i")
    println(s" | value: $j")
}
````

#### 1. Word Count using HOFs

````scala
val words = List("apple", "banana", "apple", "orange", "banana", "apple")

val wordcount = words.groupBy(identity).mapValues(_.size)
// groupBy --> will group the item in form of map[k, List[v]]
// view.mapValues(_.size) --> Count the size of the List[v] and map the value with identity

println(wordcount.toSeq)
````

````scala
// 2nd option the solve using loop
import scala.collection.mutable.Map
val result: Map[String, Int] = Map()
for (word <- words){
    result(word) = result.getOrElse(word, 0) + 1
}
println(result)
````

````scala
// using the df
val wordsDf = words.toDF("word")
wordsDf.groupBy("word").count().show() 
````

#### 2. Find Even Numbers using filter

````scala
val nums = List(1, 2, 3, 4, 5, 6)

val evanNums = nums.filter(i => i%2==0)
// val evanNums = nums.filter(_%2==0) // this is also same; here _ is anonymous func
print(evanNums)
````

#### 3. Sum of List using reduce / fold

````scala
val nums = List(1, 2, 3, 4, 5, 6)
val total = nums.reduce((a, b) => a + b)
println(total)
````

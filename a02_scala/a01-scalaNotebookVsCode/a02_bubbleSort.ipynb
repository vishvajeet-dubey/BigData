# Bubble Sort

````scala
def bubbleSort(n: Array[Int]): Array[Int] = {
    for (item <- n.indices) {
        for (j <- 0 until n.length - item - 1) {
            if (n(j) > n(j + 1)) {
                val temp = n(j)
                n(j) = n(j + 1)
                n(j + 1) = temp
            }
        }
    }
    n
}
````

````scala
bubbleSort(Array(5, 2, 9, 1, 5, 6))
````

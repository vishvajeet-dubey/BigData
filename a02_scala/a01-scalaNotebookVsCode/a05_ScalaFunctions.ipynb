## 01-Functions
* A function is a collection of statements that perform a certain task
* Scala functions are first class values. 

**Difference between Scala Functions & Methods:**   
* Function is a object which can be stored in a variable. 
* But a method always belongs to a class which has a name, signature bytecode etc. 
* Basically, you can say a method is a function which is a member of some object.

### Function Declaration & Definition
In general, function declaration & definition have 6 components:
* **def keyword:** "def" keyword is used to declare a function in Scala.
* **function_name:** It should be valid name in lower camel case. Function name in Scala can have characters like +, ~, &, --, ++, \, / etc.
* **parameter_list:** In Scala, comma-separated list of the input parameters are defined, preceded with their data type, within the enclosed parenthesis.
* **return_type:** User must mention return type of parameters while defining function and return type of a function is optional. If you don't specify any return type of a function, default return type is Unit which is equivalent to void in Java.
* = : In Scala, a user can create function with or without = (equal) operator. If the user uses it, the function will return the desired value. If he doesn't use it, the function will not return any value and will work like a subroutine.
* **Method body:** Method body is enclosed between braces { }. The code you need to be executed to perform your intended operations.


### Syntax
```scala
def function_name ([parameter_list]) : [return_type] = {
   
  // function body

}
```

````scala
def funcToSum(num1: Int, num2: Int): Int ={
    val sum = num1 + num2
    return sum // can writen as only sum that will return the value.
}

// calling the function
funcToSum(10, 5)
````

## 02-Anonymous-Functions
* An anonymous function is also known as a function literal. 
* A function which does not contain a name is known as an anonymous function.

### Syntax
```scala
(z:Int, y:Int)=> z*y
Or
(_:Int)*(_:Int)
```

In the above first syntax, `=>` is known as a transformer. The transformer is used to transform the parameter-list of the left-hand side of the symbol into a new result using the expression present on the right-hand side.   

In the above second syntax, `_` character is known as a wildcard is a shorthand way to represent a parameter who appears only once in the anonymous function.

### Anonymous Functions With Parameters
* When a function literal is instantiated in an object is known as a function value. 
* Or in other words, when an anonymous function is assigned to a variable then we can invoke that variable like a function call. 
* We can define multiple arguments in the anonymous function.

````scala
val numList = List.range(1,10)

// creating anonymous function to filter the even number
val evenNumList = numList.filter((i: Int) => i%2==0) // Anonymous function = (i: Int) => i%2==0

// we can define using _ known as wildcard
val evenNumList01 = numList.filter(_%2==0)
val oddNumList = numList.filter(_%2==1)
````

````scala
// Creating anonymous functions 
// with multiple parameters Assign 
// anonymous functions to variables 
var fun1 = (str1: String, str2: String) => str1 + str2


// An anonymous function is created 
// using _ wildcard instead of 
// variable name because str1 and 
// str2 variable appear only once 
var fun2 = (_:String) + (_:String)

println(fun1("Vishvajeet", " Dubey"))
println(fun2("Rajan", " Dubey"))
````

## 03-Recursion-in-scala
* A recursive function is simply a function that `calls itself` to solve a smaller instance of the problem.

Every distinct recursive function must have two parts to avoid running forever:
1. **Base Case:** The condition under which the function stops calling itself and returns a value.
2. **Recursive Step:** The part where the function calls itself with modified arguments to move closer to the base case.

````scala
// recusion function to calculate the factorial
// n! = n * (n-1)! where 1! = 1
// 5! = 5 * 4 * 3 * 2 * 1 = 120

def factorial(num: Int): Int={
    if (num==1) {1} // base case
    else{num * factorial(num-1)} // recursive steps
}

factorial(5)
````

````scala
// detailed view
def factorial(num: Int): Int={
    if (num==1) {1} // base case
    else{
        println(s"${num} * factorial(${num-1})")
        num * factorial(num-1)
        } // recursive steps
}

factorial(5)
````

````scala
// another example of recursion
// sum of first n arithimatic integers
def expSum(n: Int): Int = {
    if(n<=0) 0
    else{n + expSum(n-1)}
}
expSum(100000)
// Calling with a large number like 10,000:
// Since the calculation waits for the next step to finish, the JVM must 
// keep the current state (stack frame) in memory. 
// For large inputs, this fills up the stack and throws a StackOverflowError.
// To overcome this, we can use tail recursion, which passes the accumulated result 
// to the next iteration, allowing the compiler to reuse the current stack frame.
````

## 04-Tail-Recursion
* If a recursive call is the very last action performed by the function, the compiler treats it like a while loop. 
* It reuses the current stack frame instead of creating a new one.
* The recursive call must be in the "tail position." There can be no calculation pending on the result of the recursive call.
* **Standard:** `n + expSum(n - 1)` $\rightarrow$ Not tail recursive (sumation happens after the return).    
* **Tail:** `expSumTail(n - 1, n + acc)` $\rightarrow$ Tail recursive (the call is the absolute last step).   

**The @tailrec Annotation**
* Scala provides an annotation to ensure your function is actually tail-recursive. 
* If you use this annotation and the function isn't tail-recursive, the compiler will throw an error.

````scala
// let solve the above problem using tail recursion
// write a function to sum of first n arithimatic integers
def expSumTail(n: Int, accumVar: Int=0): Int = {
    // base case
    if(n<=0) {accumVar}
    
    // recursive step
    else{
        expSumTail((n-1), n + accumVar)
    }
}

expSumTail(100000) // this way we can avoid the stackOverflowError
````

````scala
val esum: (Int, Int) => Int = (n, accumulator) => if(n<=0) accumulator else{esum((n-1), n + accumulator)}
esum(100,0)
````

````scala
// using anotation tail rec
@scala.annotation.tailrec
def expSumTail(n: Int, accumVar: Int=0): Int = {
    // base case
    if(n<=0) {accumVar}
    
    // recursive step
    else{
        expSumTail((n-1), n + accumVar)
    }
}

expSumTail(10000)
````

````scala
// if you are using anotation tailrec then you have to define your function like tailrec not standard recusion else it will throw an error
@scala.annotation.tailrec
def expSum(n: Int): Int = {
    // base case
    if(n<=0) 0
    
    // recursive step
    else{
        n + expSum(n-1)
    }
}

expSumTail(100)
````

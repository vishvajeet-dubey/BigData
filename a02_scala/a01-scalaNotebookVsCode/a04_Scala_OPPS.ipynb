## 01-Class
- A class is a user-defined blueprint or prototype from which objects are created.
- A class combines the fields and methods(member function which defines actions) into a single unit.

````scala
// name of the class = SmartPhone
class SmartPhone(storeName: String) {

  println(s"This is ${storeName} store.")

  val totalNum: Int = 16
  val companyName: String = "Apple"

  // defining the function
  def display(): Unit={
    println(s"Name of company ${companyName}")
    println(s"Total number of smart phone generation ${totalNum}")
  }
}
````

* creating object of above class.  
* In a Scala notebook, we can define a singleton object to organize code, but it is not the entry point of execution. The code runs only when we call its methods, and those methods can return values.

````scala
val obj = new SmartPhone("delhi")
````

Note: as soon as object created the primary constructor, field and statement will initialized. in this case the println() statement will print as soon as object created field(totalNum and companyName) and primary constructor(storeName) will initialized.  

**lets use the method defined under the SmartPhone class using object**

````scala
// accessing the method of class using object
obj.display()
````

## 02-Inheritance
* It is the mechanism in Scala by which one class is allowed to inherit the features(fields and methods) of another class.

### 🔑 Key Points

* Scala uses the **`extends`** keyword to implement inheritance. 
* A subclass can **add new members or override existing ones** from the parent class.
* Objects of the subclass can **access parent class members**.

### 🧩 Types of Inheritance in Scala

1. **Single inheritance** — one child inherits from one parent.
2. **Multilevel inheritance** — inheritance chain (A → B → C).
3. **Hierarchical inheritance** — multiple children inherit from one parent.
4. **Multiple inheritance** — not supported with classes, but possible using **traits**.
5. **Hybrid inheritance** — combination of types, achievable through traits.



````scala
class Animal(name: String){
  println(s"This is ${name}")
  def sound(): String = {"Some sound"}
}

class Puppy extends Animal("Dog") { // Puppy class is inheriting the properties of Animal class
  override def sound(): String ={"Bark"}
  println(s"Sound is ${sound()}")
}

// creating the object of Puppy class
val obj = new Puppy()
// as soon as object created primary constructor will automatically call

// access the method of Puppy class
println(s"Dog sound is ${obj.sound}")
````

## 03-Singleton Object
* it is an object which defines the single object of class.
* it provides an entry point to your program execution.
* if you do not create singleton object then your program will compile but not return any result.

### 🔑 Key Characteristics
* Declared using the `object` keyword (not `class`).
* Automatically **instantiated once** — you don’t create it with `new`.
* Used to hold **utility methods, configuration, or shared state**.
* Can contain methods, fields, and even extend traits or classes.
* In standalone applications, an object with a `main` method acts as the program entry point.

````scala
object Logger {
  def log(msg: String): Unit = println(msg)
}

Logger.log("Hello")
````

## 05-Constructor
A **constructor** in Scala is a special mechanism used to **initialize an object when it is created**. Unlike many languages, Scala has a **primary constructor built directly into the class definition**.


### 🔑 Primary Constructor
* In Scala, the **class parameters themselves form the primary constructor**.  
* ✔ Runs automatically when the object is created  
* ✔ You don’t write a separate constructor block
* 👉 Everything inside the class body is part of the primary constructor except method

````scala
class Person(name: String, age: Int){
    println(s"Name: ${name} and age: ${age}")
}

// creating object of the Person class
val p1 = new Person("Rahul", 25)
````

### 🔑 Secondary/Auxillary constructor
Secondary constructor (also called an auxiliary constructor) provides additional ways to create an object besides the primary constructor.
**⚙️ Key Rules**
* Must be defined inside the class body
* Must call another constructor as the first statement
    * Either the primary constructor or another secondary constructor
* Used to provide multiple initialization options

It define using this keyword
```scala
def this(...) = { ... }
```
⚡It allows different parameter combinations while still ensuring the primary constructor runs first.

````scala
class Person(name: String, age: Int){
    println(s"Name: ${name} and age: ${age}")

    // Auxillary constructor
    def this(name: String){
        this(name, 18) // default value of age is 18. Calling primary constructor
    }
}

// creating object of person class
val p2 = new Person("Arun") // secondary/aux constructor will assing the default age value 18
val p3 = new Person("Mohan", 28)
````

# Filter DF practice

````python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.master("local[*]").appName("practice").getOrCreate()
spark
````

````python
flightDf = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("data/2010_summary_flight_data.csv")
flightDf.show()
````

````python
flightDf.printSchema()
````

## filter

````python
flightDf.filter(col("count")>100).show()
````

## using where condition
```python
df.where(col("col_name") operator condition)
```

````python
flightDf.where(col("count")>100).show()
````

# if else
- if else can be done by using when and otherwise like sql
- in sql it will be case and when statement and end as else codition
- here also similar use the when codition like if condition to filter the data 
- otherwise will work like else if consition not matched

````python
flightDf.withColumn("frequency", when(col("count")>500 , "cross 500")\
                    .when(col("count")> 100, "cross 100")\
                    .otherwise("not cross 100")
                    ).show()
````

````python
flightDf.withColumn("frequency", when(col("count")>500 , "cross 500")\
                    .when(col("count")> 100, "cross 100")\
                    .otherwise("not cross 100")
                    ).filter(col("frequency")=="cross 500").show()
````

## SparkSession Vs SparkContext
| Feature            | **SparkContext**                             | **SparkSession**                                   |
| ------------------ | -------------------------------------------- | -------------------------------------------------- |
| **Introduced In**  | Spark 1.x                                    | Spark 2.0                                          |
| **Purpose**        | Entry point to low-level Spark core APIs     | Unified entry point to work with Spark APIs        |
| **Supports**       | RDD API                                      | RDD, DataFrame, Dataset APIs                       |
| **SQL Support 🔥** | Needs separate `SQLContext` or `HiveContext` | Built-in support (no separate context needed)      |
| **Ease of Use**    | Verbose and more complex                     | Simplified and more user-friendly                  |
| **Usage Example**  | `val sc = new SparkContext(conf)`            | `val spark = SparkSession.builder...getOrCreate()` |
| **Replacement**    | Deprecated in many use cases                 | Recommended for modern Spark applications          |

## Importing SparkSession

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

# creating the spark session
spark = SparkSession.builder.master("local")\
                    .appName("SS_session").getOrCreate()

print(spark)

spark2 = spark.newSession()
print(spark2)


employee_df = spark.read.format("csv")\
                    .option("inferSchema","true")\
                    .option("header","true")\
                    .load("data/employee_data.csv")

employee_df.show()
```

## Creating sparkContext

```python
# using sparkContext
# creating sparkContext
sc = spark.sparkContext

rdd = sc.textFile("data/employee_data.csv")

#split each line by comma
parse_rdd = rdd.map(lambda line: line.split(","))


for row in parse_rdd.take(5):
    print(row)
```

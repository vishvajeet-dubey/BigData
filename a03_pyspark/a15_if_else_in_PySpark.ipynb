# If Else in PySpark
- **Important Question**
    - Case when in spark SQL
    - when otherwise in PySpark
    - How to deal with null value ?
    - CAse when | when otherwise with multiple AND, OR condition

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

spark = SparkSession.builder.appName("Conditional_test")\
        .master("local[*]")\
        .getOrCreate()

spark
```

```python
flightDf = spark.read.format("csv")\
        .option("header","true")\
        .option("inferSchema","false")\
        .load("data/2010_summary_flight_data.csv")
```

```python
flightDf.show(5)
```

### Flight count is greater than 50
- if greater than or equal to 50 --> good booking
- if less than 50 --> bad booking

```python
flightDf = flightDf.withColumn("booking", when(col("count") >= 100, "Cross Century")\
                    .when(col("count") >= 50, "good booking")\
                    .otherwise("bad booking"))
```

```python
# filter top 5 good booking flights
flightDf.filter(col("booking")=="good booking").orderBy(col("count").desc()).show(5)
```

# Bubble Sort
### Definition
* Bubble Sort is a simple comparison-based sorting algorithm.
* It repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order.
* This process continues until the list is sorted.
* It is named Bubble Sort because the larger elements move to the end while the smaller elements rise to the top, like bubbles in water.


### Algorithm Steps
1. Start from the beginning of the list/array.
2. Compare the current element with the next one.
3. If the current element is greater, swap them.
4. Move to the next pair and repeat.
5. After one full pass, the largest element bubbles up to the end.
6. Repeat the process for remaining elements until no swaps are needed.


### Example
```cmd
Sorting [5, 3, 8, 1, 2]
----------------------------
Pass 1: [3, 5, 1, 2, 8]
Pass 2: [3, 1, 2, 5, 8]
Pass 3: [1, 2, 3, 5, 8]
Pass 4: [1, 2, 3, 5, 8]  (already sorted)
```

### Time Complexity
* Best case: O(n) (when already sorted, if optimized with a swap flag).
* Average case: O(n²).
* Worst case: O(n²) (when sorted in reverse order).


### Space Complexity
* O(1) → Bubble sort sorts the array in-place (no extra memory required).


### Properties
* Stable → Equal elements remain in the same order.
* Adaptive → Can stop early if the list becomes sorted before all passes.
* Simple but inefficient for large datasets.

````python
sampleList: list = [42, 7, 19, 3, 88, 56, 21, 73, 15, 90, 12, 67, 2, 34, 81, 5, 49, 28, 61, 10]

# sorted(sampleList)

def bubbleSort(aList: list) -> list:
    for i in range(len(aList)-1):
        for j in range(0, len(aList)-i -1):
            if (aList[j] > aList[j+1]):
                # swap
                (aList[j], aList[j+1]) = (aList[j+1], aList[j])
    return aList

bubbleSort(sampleList)
````

### Que - 01
#### Python Program to check prime number
- **Prime number** - a prime number is number which can only devided one or itself
- ex. - 2, 3, 5, 7, 13, 17, 19, 23, 29, 31, 37, 41, 47...

````python
# tried from my side
def primeNumber(num: int) -> str :
    factor = 0
    if num<=1:
        return f"{num} is not prime number"
    else:
        for i in range(1, num+1):
            if num%i == 0:
                factor += 1
        if factor > 2 :
            return f"{num} is not prime number"
        else:
            return f"{num} is prime number"

primeNumber(31)
````

````python
# otpimized way
def primeNumber(num: int) -> str :
    if num<=1:
        return f"{num} is not prime number"
    else:
        for i in range(2, num):
            if num%i == 0:
                return f"{num} is not prime number"
        return f"{num} is prime number"

primeNumber(48)
````

### Que 02
#### Find factorial of number
- it a function that multiple a number by every number below it till 1.
- 5! = 5 * 4 * 3 * 2 * 1 

````python
# solved by me
def fact(num: int) -> int: # FYI python function not validate the ruturn type when you u provide explicitly. it is just as hint for reading devs
    if num >= 1:
        result = 1
        for n in range(2, num + 1):
            result = result * n
        return result
    elif num==0 or num == 1:
        return 1
    else:
        return "Factorial not possible"
fact(10)
````

````python
# factorial using the recursion -> optimized 
def fact(num: int) -> int:
    if num < 0:
        return "factorial not possible"
    else:
        if num == 0 or num == 1:
            return 1
        else:
            return num * fact(num -1)
fact(5)
````

### Que 03
#### Fibonacci number
- The Fibonacci sequence is a infinite series of numbers where every value is found by adding the two previous numbers together.
```python
f(n) = f(n-1) + f(n-2)
where,
f(0) = 0
f(1) = 1
```

````python
def febonacci(num: int) -> int:
    if num == 0:
        return 0
    if num == 1:
        return 1
    return febonacci(num-1) + febonacci(num-2)

febonacci(7)
````

### Que 04
#### Sum of square of natural number 
- ex 4 => 1^2 + 2^2 + 3^2 + 4^2 = 30

````python
# using function and recursion
def sq(num: int) -> int : return num*num # same thing can be write using lambda function ``squ = lambda num : num * num``
def sumOfSquareNum(num: int) -> int : 
    if num == 0: 
        return 0
    else : 
        return sq(num) + sumOfSquareNum(num-1)
sumOfSquareNum(1000)
````

````python
# using loop
def sumOfSquareNum(num: int) -> int : 
    sqsum = 0
    for i in range(1, num + 1):
        sqsum += i*i
    return sqsum
sumOfSquareNum(1000)
````

### Que 05
#### Armstrong number
- An Armstrong number (also known as a narcissistic number) is a positive integer that is equal to the sum of its own digits each raised to the power of the number of digits.
- The most famous example is 153 because it has 3 digits, and 1³ + 5³ + 3³ = 1 + 125 + 27 = 153.

````python
def power(num: int, exp: int)-> int : return num**exp # defined power manually
def armstrongNum(num: int) -> bool :
    length = len(str(num))
    total = 0
    for i in str(num):
        total += power(int(i), length)
    if num == total:
        return True
    else:
        return False
    # return length
armstrongNum(153)
````

````python
for i in range (1, 1000):
    if armstrongNum(i)==True:
        print(i)
````

#### Sum of array without using inbould function

````python
def sumOfArray(arr: list) -> int:
    sum = 0
    for i in arr:
        sum += i
    return sum

sumOfArray([1,2,3,4,5,6])
````

#### Python program to find the larget list

````python
def largetVal(arr: list) -> int:
    larget = 0
    for i in arr:
        if i>larget:
            larget=i
    return i
largetVal([23,2,24,34,34,354,3445])
````

### Write a program to reverse in given string.

````python
def reverseString(sentance: str) -> str:
    sList = sentance.split(" ")
    return " ".join(sList[::-1])

reverseString("this program is for testing")
````

### Que 08
#### find the word count in string
- input - "this is word word"
- this: 1, is: 1, word: 2

````python
def wordCount(sen: str) -> dict :
    wordList = sen.split(" ")
    wc = dict()
    for i in wordList:
        wc[i]=wordList.count(i)
    return wc

wordCount("this is word word")
````

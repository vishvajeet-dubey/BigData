## SPARK PRACTICE:-BASIC
### Create spark session 

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark = SparkSession.builder.master("local[*]")\
        .appName("sparkPractice") \
        .getOrCreate()

spark
```

### 1. Read a dataset – basic way 

```python
df=spark.read.csv("data/emp_data.csv")
df.show()
```

### 2. Read dataset with column name.

```python
df = spark.read.format("csv")\
    .option("header", "true")\
    .load("data/emp_data.csv")
df.show()
```

### 3. Save dataset in a variable

```python
pyspark_df = spark.read.format("csv")\
            .option("header", "true")\
            .option("inferschema", "true")\
            .load("data/employee_data.csv")
pyspark_df.show()
```

### 19. Delete row having null value

```python
dfWithNullRow = spark.read.format("csv").option("header","true").option("inferSchema", "true").load("data/employee_data_a12.csv")
dfWithNullRow.show()
```

```python
dfWithNullRow.na.drop().show()
```

### 20. Delete row if all values are null

```python
dfWithNullRow.na.drop(how="all").show()
```

### 21. Delete row if 2 column are null (#total number of column is 6)

```python
dfWithNullRow.na.drop(how="any", thresh=5).show()
```

### 22. Delete record if there is null value in specific column

```python
dfWithNullRow.show()
dfWithNullRow.na.drop(how="any", subset=['salary']).show()
dfWithNullRow.na.drop(how="any", subset=['salary', 'address']).show()
```

### 23. Fill 'not avl' in null value ---only works for string data--
💡 Note: if integer then fill 0

```python
dfWithNullRow.na.fill('not avl').show()
```

```python
# for int and string both
dfWithNullRow.na.fill(0).na.fill('not avl').show()
```

### 24. Fill 'not avl' in null value of specific column(eg: address)

```python
dfWithNullRow.na.fill('not avl', 'address').show()
```

### 25. Fill 'not avl' in null value of multiple column(eg: name and address)

```python
dfWithNullRow.na.fill('not avl', ['name', 'address']).show()
```

### 26. Change datatype of "salary" column to double
💡 Note: for this, `col` should be imported from pyspark.sql.functions

```python
dfWithNullRowNew = dfWithNullRow.withColumn("salary", col("salary").cast("double"))
dfWithNullRowNew.printSchema()
```

# Parquet
- Parquet is Columnar based file format
- Structure file format
- in binary form
- It is commonly used in big data framework like Appace spark, Hadoop and Hive

##### Key Features:
- **Columnar Storage**
  - Store data by `columns` rathr than rows.
  - Enable reading only the `required columns`, which speeds up query performace.
- **Compression and Encoding**
  - Support various compression algorithms(e.g., Snappy, Gzip, Zstandard)
  - Compression reduces storage usage and improces I/O performace.
- **Schema Evolution**
  - Allows changes to the schema over time without distrupting the stored data.
- Predicate Pushdown
  - Support filtering operations directly at the file level.
  - Reduces unnecessary data read by filtering data as it's being scanned.

#### What is columnar and row based file format
**Columnar based file format**  
- Data is stored in a columnar format, meaning that all records for a particular column are stored together on disk.   

For example: Below is the sample table   

| a   | b   | c   |
|-----|-----|-----|
| a1  | b1  | c1  |
| a2  | b2  | c2  |
| a3  | b3  | c3  |

Data will store together for column on disk   

Record on disk   
|a|a1|a2|a3|b|b1|b2|b3|c|c1|c2|c3|
|---|---|---|---|---|---|---|---|---|---|---|---|


**Row based file format**
- Entire rows stored together

For example: Below is the sample table   

| a   | b   | c   |
|-----|-----|-----|
| a1  | b1  | c1  |
| a2  | b2  | c2  |
| a3  | b3  | c3  |

Data will store together for row on disk   

Record on disk   
| a   | b   | c   | a1  | b1  | c1  |a2  | b2  | c2  | a3  | b3  | c3  |
|---|---|---|---|---|---|---|---|---|---|---|---|

------------


🔥 **Row-Based vs. Columnar Storage**
| **Feature**        | **Row-Based Storage**                | **Columnar Storage**                  |
|--------------------|------------------------------------|-------------------------------------|
| **Storage Layout**  | Entire rows stored together         | Columns stored together              |
| **Best for**        | OLTP (frequent row retrieval)       | OLAP (large-scale data analysis)     |
| **Query Efficiency**| Slower for analytical queries       | Faster for analytical queries        |
| **Write Performance**| Faster for inserts/updates         | Slower for individual row writes     |
| **Compression**     | Less effective                      | More effective due to column similarity |

# Reading parquet file
- `spark.read.parquet("path/file_name.parquet")`
- No need to pass inferSchema and other option in parquet file as it already store column level information.

```
df = spark.read.parquet("/FileStore/tables/2010_flight_data.gz")
df.show(5)
```

# Unique

1. How to find unique rows ?
2. How to drop duplicates rows ?
3. How to sort the data in ascending and descending order ?
4. 1 simple question of pyspark.

````python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark = SparkSession.builder.master("local[*]").appName("unique").getOrCreate()
spark
````

````python
dataManager=[(10 ,'Anil',50000, 18),
(11 ,'Vikas',75000,  16),
(12 ,'Nisha',40000,  18),
(13 ,'Nidhi',60000,  17),
(14 ,'Priya',80000,  18),
(15 ,'Mohit',45000,  18),
(16 ,'Rajesh',90000, 10),
(17 ,'Raman',55000, 16),
(18 ,'Sam',65000,   17),
(15 ,'Mohit',45000,  18),
(13 ,'Nidhi',60000,  17),      
(14 ,'Priya',90000,  18),  
(18 ,'Sam',65000,   17)
]
schemaManager = ["id", "name", "sal", "manager_id"]

managerDf = spark.createDataFrame(data=dataManager, schema=schemaManager)
````

````python
managerDf.show()
````

````python
# distinct() - it will return unique records
# no parameters and call it directly like below
managerDf.distinct().show()
````

````python
# check how many dups record
managerDf.count() - managerDf.distinct().count()
````

````python
# check the unique record on specific column
managerDf.select("id","name").distinct().show()
````

````python
#drop duplicate
managerDfUnique = managerDf.drop_duplicates(("id", "name", "sal", "manager_id")).show()
````

# Sort

````python
managerDf.sort(col("sal").desc(), col("name").desc()).show()
````

# LeetCode quetions

```txt
Table: Customer

+-------------+---------+
| Column Name | Type    |
+-------------+---------+
| id          | int     |
| name        | varchar |
| referee_id  | int     |
+-------------+---------+
```

In SQL, id is the primary key column for this table.
Each row of this table indicates the id of a customer, their name, and the id of the customer who referred them.
 

Find the names of the customer that are either: 
1. **referred by any customer** with id != 2.
2. **not referred** by any customer.
- Return the result table in any order.

The result format is in the following example.

 
```txt
Example 1:

Input: 
Customer table:
+----+------+------------+
| id | name | referee_id |
+----+------+------------+
| 1  | Will | None       |
| 2  | Jane | None       |
| 3  | Alex | 2          |
| 4  | Bill | None       |
| 5  | Zack | 1          |
| 6  | Mark | 2          |
+----+------+------------+
Output: 
+------+
| name |
+------+
| Will |
| Jane |
| Bill |
| Zack |
+------+
```

[Click here](https://leetcode.com/problems/find-customer-referee/?envType=study-plan-v2&envId=top-sql-50) to open the quetion in leetCode.

````python
custData = [(1,"Will",None),
(2,"Jane",None),
(3,"Alex",2),
(4,"Bill",None),
(5,"Zack",1),
(6,"Mark",2)]

custSchema = ["id", "name", "referee_id"]

custDf = spark.createDataFrame(data=custData, schema=custSchema)
custDf.show()
````

````python
# 1. **referred by any customer** with id != 2.
# & --> and operator
# | --> or operator
# solution 
custDf.select(col("name")).where((col("referee_id").isNull()) | (col("referee_id")!=2)).show()
````

````python
# 2. **not referred** by any customer.
# Not referred by any customer means column referee_id will be null
# solution
custDf.select(col("name")).where(col("referee_id").isNull()).show()
````
